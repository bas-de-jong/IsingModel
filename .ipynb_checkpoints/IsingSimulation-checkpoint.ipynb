<h2>Definitions of Functions</h2>

```python
def initialize(L,ordered=False):
    """
    Initializes array of size LxL in ordered state: all spins one, 
    or unordered state: spins 1 or -1 with 50% probability.
    """
    if ordered:
        spins = np.ones((L,L),dtype=int)
    else:
        spins = 2*np.random.random_integers(0,1,(L,L))-1
    return spins

def simulate_step_Metropolis(spins,J,T):
    """
    Makes a single Metropolis step using 'red-black' updating.
    Args: 
        spins: array of size LxL with integers 1 and -1
        J: Interaction between spins, float
        T: Temperature, float
    Returns:
        spins: spins array modified after one step using the Metropolis method.
        magnetization: magnetization of the lattice.
        energy_tot: total energy of the current spin configuration.
    """
    total_neighbours = np.roll(spins,1,0) + np.roll(spins, -1, 0) + np.roll(spins, -1, 1) + np.roll(spins, 1, 1)
    energy_spins = -total_neighbours*spins*J
    energy_diff = -2*energy_spins
    energy_tot = np.sum(energy_spins)
    
    boltz_weights = np.exp(-energy_diff*(1/T))
    random_weights = np.random.random(spins.shape)
    
    flips = np.ones(spins.shape,dtype=int)
    flips[random_weights < boltz_weights] = -1
    new_spins = spins*flips
    
    #The black sites
    spins[0:L:2,0:L:2] = new_spins[0:L:2,0:L:2]
    spins[1:L:2,1:L:2] = new_spins[1:L:2,1:L:2]
    
    total_neighbours = np.roll(spins,1,0) + np.roll(spins, -1, 0) + np.roll(spins, -1, 1) + np.roll(spins, 1, 1)
    energy_spins = -total_neighbours*spins*J
    energy_diff = -2*energy_spins
    energy_tot = np.sum(energy_spins)/2
    
    boltz_weights = np.exp(-energy_diff*(1/T))
    random_weights = np.random.random(spins.shape)
    
    flips = np.ones(spins.shape,dtype=int)
    flips[random_weights < boltz_weights] = -1
    new_spins = spins*flips
    
    #The red sites
    spins[1:L:2,0:L:2] = new_spins[1:L:2,0:L:2]
    spins[0:L:2,1:L:2] = new_spins[0:L:2,1:L:2]
    
    magnetization = np.sum(spins)/L**2
    return spins, magnetization, energy_tot

def simulate_step_SW(spins,J,T,L):
    """
    Makes a single step using the Swendsen-Wang method. 
    Args: 
        spins: array of size LxL with integers 1 and -1.
        J: Interaction between spins, float.
        T: Temperature, float.
    Returns:
        spins: spins array modified after one step using the Swendsen-Wang method.
        labels: array of size LxL with integers indicating what cluster each spin is in.
        magnetization: magnetization of the lattice.
        energy_tot: total energy of the current spin configuration.
    """
    L = len(spins[0,:])
    link_above = np.roll(spins, 1, 0)*spins
    link_left = np.roll(spins, 1, 1)*spins    
    P_above = np.random.random((L,L))
    P_left = np.random.random((L,L))
    link_P = 1-np.exp(-2*J/T)
    
    link_above[P_above > link_P] = -1
    link_left[P_left > link_P] = -1
    
    label = np.reshape(np.arange(L**2,dtype=int),(L,L))
       
    for i in range(L):
        for j in range(L):
            if link_above[i,j] == 1 and link_left[i,j] == 1:
                label[label == label[(i-1)%L,j]] = label[i,(j-1)%L]
                label[i,j] = label[i,(j-1)%L]
            elif link_above[i,j] == -1 and link_left[i,j] == 1:
                label[i,j] = label[i,(j-1)%L]
            elif link_above[i,j] == 1  and link_left[i,j] == -1:
                label[i,j] = label[(i-1)%L,j]
                
    for i in range(L**2):
        x, y = np.where(label == i)
        spins[x,y] = spins[x,y]*(2*np.random.random_integers(0,1,1)-1)
    
    total_neighbours = np.roll(spins,1,0) + np.roll(spins, -1, 0) + np.roll(spins, -1, 1) + np.roll(spins, 1, 1)
    energy_tot = np.sum(-total_neighbours*spins*J)
    magnetization = np.sum(spins)/L**2
    
    return spins, label, magnetization, energy_tot
```

```python
%%cython
import cython
import numpy as np
cimport numpy as np
cimport cython
@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)

def simulate_step_SW_C(np.ndarray[np.int_t, ndim=2, mode='c'] spins,
                                   np.ndarray[np.int_t, ndim=1, mode='c'] randomints,
                                   float J, float T):
    """
    Makes a single step using the Swendsen-Wang method. 
    Args: 
        spins: array of size LxL with integers 1 and -1.
        randomint: array of size L^2x1 with random integers 1 and -1. Used to improve speed.
        J: Interaction between spins, float.
        T: Temperature, float.
    Returns:
        spins: spins array modified after one step using the Swendsen-Wang method.
        labels: array of size LxL with integers indicating what cluster each spin is in.
        magnetization: magnetization of the lattice.
        energy_tot: total energy of the current spin configuration.
    """
    #Static variable declarations
    cdef int L = len(spins[0,:])
    cdef np.ndarray[np.int_t, ndim=2, mode='c'] link_above = np.roll(spins, 1, 0)*spins
    cdef np.ndarray[np.int_t, ndim=2, mode='c'] link_left = np.roll(spins, 1, 1)*spins
    cdef np.ndarray[np.int_t, ndim=2, mode='c'] label = np.reshape(np.arange(L**2,dtype=int),(L,L))
    cdef np.ndarray[np.int_t, ndim=2, mode='c'] spinsout = np.zeros((L,L), dtype=int)
    cdef np.ndarray[np.double_t, ndim=2, mode='c'] P_above = np.random.random((L,L))
    cdef np.ndarray[np.double_t, ndim=2, mode='c'] P_left = np.random.random((L,L))
    cdef double link_P = 1-np.exp(-2*J/T), magnetization
    cdef int energy_tot
    cdef unsigned int i, j, above, left, k, l
    
    link_above[P_above > link_P] = -1
    link_left[P_left > link_P] = -1
    
    for i in range(L):
        for j in range(L):
            above = <unsigned int>((i-1)%L)
            left = <unsigned int>((j-1)%L)
            if link_above[i,j] == 1 and link_left[i,j] == 1:
                for k in range(L):
                    for l in range(L):
                        if label[k,l] == label[i,left]:
                            label[k,l] = label[above,j]
                            spinsout[k,l] = randomints[label[above,j]]
                label[i,j] = label[above,j]
                spinsout[i,j] = randomints[label[above,j]]
            elif link_above[i,j] == -1 and link_left[i,j] == 1:
                label[i,j] = label[i,left]
                spinsout[i,j] = randomints[label[i,left]]
            elif link_above[i,j] == 1  and link_left[i,j] == -1:
                label[i,j] = label[above,j]
                spinsout[i,j] = randomints[label[above,j]]
            else:
                spinsout[i,j] = randomints[label[i,j]]

    total_neighbours = np.roll(spinsout,1,0) + np.roll(spinsout, -1, 0) + np.roll(spinsout, -1, 1) + np.roll(spinsout, 1, 1)
    energy_tot = np.sum(-total_neighbours*spinsout*J)
    magnetization = np.sum(spinsout)/L**2
    
    return spinsout, label, magnetization, energy_tot
```

<h3>Simulation Functions</h3>

```python
def simulation_Metropolis(T_vec,L,J,steps):
    """
    Simulates the system for a range of temperatures and a number of timesteps.
    Args:
        T_vec: A vector containing the temperatures at which the system must be simulated.
        L: Size of the system.
        J: Interaction between spins.
        steps: number of steps that the system will take, int.
    Returns:
        energy_time: An array of size len(T_vec) x (steps+equilibrium) which contains the energy at each step.
        mag_time: An array of size len(T_vec) x (steps+equilibrium) which contains the magnetization at each step.
    """
    #initialize observables
    energy_time = np.zeros((len(T_vec),steps))
    mag_time = np.zeros((len(T_vec),steps))
    
    for t in range(len(T_vec)):
        T = T_vec[t]
        spins = initialize(L,True)
    
        for i in range(steps):
            spins, magnetization, energy = simulate_step_Metropolis(spins,J,T)
            energy_time[t,i] = energy
            mag_time[t,i] = magnetization
            
        print(t, end=" ")    
    return energy_time, mag_time
```

```python
def simulation_SW_C(T_vec,L,J,steps):
    """
    Simulates the system for a range of temperatures and a number of timesteps.
    Args:
        T_vec: A vector containing the temperatures at which the system must be simulated.
        L: Size of the system.
        J: Interaction between spins.
        steps: number of steps that the system will take, int.
    Returns:
        energy_time: An array of size len(T_vec) x (steps+equilibrium) which contains the energy at each step.
        mag_time: An array of size len(T_vec) x (steps+equilibrium) which contains the magnetization at each step.
    """
    #initialize observables
    energy_time = np.zeros((len(T_vec),steps))
    mag_time = np.zeros((len(T_vec),steps))
    
    for t in range(len(T_vec)):
        T = T_vec[t]
        spins = initialize(L,True)
    
        for i in range(steps):
            randomints = 2*np.random.random_integers(0,1,L*L)-1
            spinsout, labels, magnetization, energy = simulate_step_SW_C(spins,randomints,J,T)
            energy_time[t,i] = energy
            mag_time[t,i] = magnetization
            spins = spinsout
            
        print(t, end=" ")    
    return energy_time, mag_time
```

```python
def comp_thermo_quantities(T_vec,L,J,energy_time,mag_time,steps,equilibrium,block_len):
    """
    Computes the thermodynamic quantities from the simulated data.
    Args:
        T_vec: A vector containing the temperatures at which the system must be simulated.
        L: Size of the system.
        J: Interaction between spins.
        steps: number of steps that the system will take, int.
        equilibrium: number of steps for the system to reach equilibrium, int.
        energy_time: An array of size len(T_vec) x (steps+equilibrium) which contains the energy at each step.
        mag_time: An array of size len(T_vec) x (steps+equilibrium) which contains the magnetization at each step.
    Returns:
        E_mean: Mean energy at each temperature in T_vec.
        E_std: Error of E_mean.
        Cv_mean: Mean specific heat at each temperature in T_vec.
        C_std: Error of Cv_mean.
        mag_mean: Mean magnetization at each temperature in T_vec.
        mag_std: Error of mag_mean.
        mag_rms_mean: Mean RMS value of the magnetization at each temperature in T_vec.
        mag_rms_std: Error of mag_rms_std.
        binCum_mean: Mean 4th order Binder Cumulant at each temperature in T_vec.
        binCum_std: Error of binCum_std.
        Chi_mean: Mean magnetic susceptibility at each temperature in T_vec.
        Chi_std: Error of Chi_mean.        
    """
    E_mean = np.zeros(len(T_vec))
    E_std = np.zeros(len(T_vec))
    
    Cv_mean = np.zeros(len(T_vec))
    Cv_std = np.zeros(len(T_vec))
    
    mag_std = np.zeros(len(T_vec))
    mag_mean = np.zeros(len(T_vec))
    
    mag_rms_std = np.zeros(len(T_vec))
    mag_rms_mean = np.zeros(len(T_vec))
    
    binCum_mean = np.zeros(len(T_vec))
    binCum_std = np.zeros(len(T_vec))
    
    Chi_mean = np.zeros(len(T_vec))
    Chi_std = np.zeros(len(T_vec))
    
    for t in range(len(T_vec)):
        T = T_vec[t]
              
        #Blocking energy
        energy_t = energy_time[t,equilibrium:steps-((steps-equilibrium)%block_len)]
        blocks_e = np.reshape(energy_t,(block_len,len(energy_t)/block_len))
         
        #Use blocked energy for E/N
        E_mean[t] = np.mean(np.mean(blocks_e/L**2))
        E_std[t] = np.std(np.mean(blocks_e/L**2,1))
       
        #Use blocked energy for Cv    
        Cv = (L*T)**(-2)*(np.mean(blocks_e**2,1)-np.mean(blocks_e,1)**2)
        Cv_std[t] = np.std(Cv)
        Cv_mean[t] = np.mean(Cv)
        
        #Blocking magnetization
        mag_t = mag_time[t,equilibrium:steps-((steps-equilibrium)%block_len)]
        blocks_mag = np.reshape(mag_t,(block_len,len(energy_t)/block_len))
        
        #Use blocked magnetization for error magnetization
        mag_mean[t] = np.mean(np.mean(blocks_mag))
        mag_std[t] = np.std(np.mean(blocks_mag,1))
        
        #Use blocked magnetization for susceptibility
        Chi = (L**2/T)*(np.mean(blocks_mag**2,1)-np.mean(np.abs(blocks_mag),1)**2)
        Chi_mean[t] = np.mean(Chi)
        Chi_std[t] = np.std(Chi)
        
        #Compute the RMS of magnetization
        mag_rms = (np.mean(blocks_mag**2,1))**(1/2)
        mag_rms_mean[t] = np.mean(mag_rms)
        mag_rms_std[t] = np.std(mag_rms)
        
        #Compute the binder cumulant
        binCum = 1 - np.mean(blocks_mag**4,1)/(3*np.mean(blocks_mag**2,1)**2)
        binCum_mean[t] = np.mean(binCum)
        binCum_std[t] = np.std(binCum)
        
    return E_mean, E_std, Cv_mean, Cv_std, mag_mean, mag_std, Chi_mean, Chi_std, mag_rms_mean, mag_rms_std, binCum_mean, binCum_std
```

<h2>Simulations using Swendsen-Wang</h2>

Determine a range of thermodynamic properties and saving the data. By acquiring these properties for a range of different lattice sizes it is possible to approximate the critical exponents using finite size scaling.

```python
#Defines a set of vectors which can be used to gather the data for different system sizes.
#To acquire accurate data the number of steps in steps_vector and block_len_vector must be on the order 10000 and 1000.
#However, this does require a long time to compute.
J = 1
T_vec = np.linspace(1,5,100)

L_vector = np.array([5, 10, 15, 20, 25, 30, 40])      
equilibrium = 100
steps_vector = 1000*np.array([5, 5, 5, 5, 5, 10, 10])+equilibrium
block_len_vector = 100*np.array([1, 1, 1, 1, 1, 2, 2])
```

```python
# Select a value of L from L_vector
select = 0
L = L_vector[select]
steps = steps_vector[select]
block_len = block_len_vector[select]
```

```python
#Simulates the system for a range of temperatures and determines the physical quantities
energy_time, mag_time = simulation_SW_C(T_vec,L,J,steps)
E_mean, E_std, Cv_mean, Cv_std, mag_mean, mag_std, Chi_mean, Chi_std, mag_rms_mean, mag_rms_std, binCum_mean, binCum_std \
= comp_thermo_quantities(T_vec,L,J,energy_time,mag_time,steps,equilibrium,block_len)
```

<h2>Determining the critical exponents</h2>

<h3>Load Variables for determining $\beta$ and $\gamma$</h3>

```python
L_vector = np.array([5, 10, 15, 20, 25, 30, 40])
selection = np.array([0, 1, 2, 3, 4]) #Select which points to load from L_vector
simtype = 'SWdata' #Select the simulation type: 'SW' or 'Metropolis' or 'SWdata' (SWdata is already filled)

Chi_max = np.zeros(len(selection))
L_vec = np.zeros(len(selection))
M_rms = np.zeros(len(selection))

for i in range(len(selection)):
    j = selection[i]
    L = L_vector[j]
    T_vec = np.load('.\Data\%s\Temp_L=%d.npy' %(simtype,L))
    Chi_mean = np.load('.\Data\%s\Chi_mean_L=%d.npy' %(simtype,L))
    mag_rms_mean = np.load('.\Data\%s\magnet_rms_mean_L=%d.npy' %(simtype,L))
    
    Chi_max[i] = max(Chi_mean)
    
    M_rms[i] = mag_rms_mean[np.argmin(np.abs(T_vec-2.26918))]
    
    L_vec[i] = L
```

<h4>Gamma, $\gamma$</h4>

```python
p  = np.polyfit(np.log(L_vec),np.log(Chi_max),1)
gamma = p[0]
b = p[1]

plt.figure()
plt.plot(L_vec,Chi_max,'o')
plt.plot(np.linspace(2,60,20),np.linspace(2,60,20)**gamma*np.exp(b), ls='--',color='blue')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('L')
plt.ylabel(r'$\chi_{max}$')
plt.show()
```

<h4>Beta, $\beta$</h4>

```python
p = np.polyfit(np.log(L_vec),np.log(M_rms),1)
beta = -p[0]
b = p[1]

plt.figure()
plt.plot(L_vec,M_rms,'o')
plt.plot(np.linspace(2,60,20),np.linspace(2,60,20)**(-beta)*np.exp(b), ls='--',color='blue')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('L')
plt.ylabel(r'$M_{rms}$')
plt.show()
```

<h2>Determining the performance of Metropolis and SW methods<\h3>

```python
steps = 20

L_vec = np.zeros(steps)
t_vec = np.zeros(steps)
for i in range(steps):
    L = 40 + 5*i
    spins = 2*np.random.random_integers(0,1,(L,L))-1
    randomints = 2*np.random.random_integers(0,1,L*L)-1
    timing = %timeit -o simulate_step_SW_C(spins,randomints,1,2.5);
    L_vec[i] = L
    t_vec[i] = timing.best
```

```python
p = np.polyfit(np.log(L_vec),np.log(t_vec*1e3),1)
```

```python
steps = 25

L_vec_2 = np.zeros(steps)
t_vec_2 = np.zeros(steps)
for i in range(steps):
    L = 40 + 5*i
    spins = 2*np.random.random_integers(0,1,(L,L))-1
    timing = %timeit -o simulate_step_Metropolis(spins,1,2.5);
    L_vec_2[i] = L
    t_vec_2[i] = timing.best
```

```python
p_2 = np.polyfit(np.log(L_vec_2),np.log(t_vec_2*1e3),1)
```

```python
plt.figure()
plt.plot(L_vec,t_vec*1e3,marker='x',label='Swendsen-Wang')
plt.plot(np.linspace(15,250,20),np.linspace(15,250,20)**p[0]*np.exp(p[1]),color='blue', ls='--')
plt.plot(L_vec_2,t_vec_2*1e3,marker='x',label='Metropolis',color='red')
plt.plot(np.linspace(15,250,20),np.linspace(15,250,20)**p_2[0]*np.exp(p_2[1]), color='red',ls ='--')
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'size, $L$')
plt.ylabel(r'time (ms)')
plt.legend(loc='best')
#plt.savefig('.\Figures\Performance')
plt.show()
```
